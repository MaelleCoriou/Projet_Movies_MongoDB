````python
from IPython.display import Image
Image("Pics/Mongo-exe.png")
````

---

#### Créer le répertoire \data\db à la racine de votre ordinateur

```python
cd C:\
md "\data\db"
```

---

#### Créer le chemin d'enregistrement des bases de données Mongo : --dbpath avec invite Windows :

```py
C:\Users\utilisateur>"C:\Program Files\MongoDB\Server\4.4\bin\mongod.exe" --dbpath="C:\data\db"
```
